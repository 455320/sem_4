Dataset:
| Age Group | Gender | Favorite Fruit |
|-----------|--------|----------------|
| 10-15     | Male   | Apple          |
| 10-15     | Female | Banana         |
| 16-20     | Male   | Orange         |
| 16-20     | Female | Mango          |
| 21-25     | Male   | Banana         |
| 21-25     | Female | Mango          |



```python
gender = int(input("enter your gender  \n 1 for male  \n 2 for female \n your input : "))
if gender not in [1,2]:
    print("you entered wrong number retry ")
else:
    if gender == 1 :
        gender = "m" 
    elif gender == 2 :
        gender = "f"
```

```python
age = int(input("enter your age :"))
if age > 25 :
    print("age not correct ")
elif age > 10 and age < 15 :
    if gender ==  "m":
        print("apple")
    if gender == "f":
        print("banana")
elif age > 16 and age < 20 :
    if gender ==  "m":
        print("orange")
    if gender == "f":
        print("mango")
elif age > 20 and age < 25 :
    if gender ==  "m":
        print("banana")
    if gender == "f":
        print("mango")
```
